## 1. Exploring Common Words and Phrases: <a class="anchor" id="first-bullet"></a>

### Find the Most Commonly Used Words Across All Texts

```python
#create one large array, combining all texts
tmp = df['tokenized_txt_array'].to_numpy()
arr_of_all_texts = np.concatenate(tmp, axis=0)

#create a frequency distribution using the combined text
fdist = FreqDist(arr_of_all_texts)

#create a word, frequency dataframe (collective texts)
df_fdist = pd.DataFrame.from_dict(fdist, orient='index')
df_fdist.columns = ['Frequency']
df_fdist.index.name = 'word'
df_fdist.sort_values(by=['Frequency'], inplace=True, ascending=False)

#create our word cloud
string_of_all_texts = " ".join(arr_of_all_texts)
wordcloud = WordCloud().generate(string_of_all_texts)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.savefig('../figs/all_text_word_cloud.png')
plt.show()

#export the most important words as a csv for manual checking for words to do with world vs. humans
df_fdist_copy = df_fdist.copy()
df_fdist_copy.reset_index(inplace=True)
df_fdist_copy.to_csv('../output/word_frequencies.csv')
```

As we can see there are many interesting insights. Common sentence words such as "will", "one", "saw", "thing", and "must" are frequently used. In terms of topics, a lot of topics that arise involve "man", "nature", "time", "world", and "object" for instance, which is what I would have intuitively thought philosophers would be talking about!

### Find the Most Commonly Used Words By Author

Do certain authors write about the same things as others? Or do they have a particular area of interest? 

Let's find out!

```python
#create a list of unique authors
unique_authors = df.author.unique()

#create a dictionary to store each authors dataframe
authors_dataframe_dictionary = {author : pd.DataFrame for author in unique_authors}

#create the dataframes and assign them to their author
for key in authors_dataframe_dictionary.keys():
    authors_dataframe_dictionary[key] = df[:][df.author == key]

authors_text_dictionary = {author : str for author in unique_authors}

#create a table to store each author and their collective text string
for key in authors_dataframe_dictionary.keys():
    author_name = key
    author_df = authors_dataframe_dictionary[key]
    tmp = author_df['tokenized_txt_array'].to_numpy()
    arr_of_all_texts = np.concatenate(tmp, axis=0)
    string_of_all_texts = " ".join(arr_of_all_texts)
    authors_text_dictionary[author_name] = string_of_all_texts
    
fig = plt.figure()
fig.set_size_inches(20, 20)

for i in range(len(authors_text_dictionary.keys())):
    ax = fig.add_subplot(6,6,i+1)
    author = list(authors_text_dictionary)[i]
    ax.title.set_text(author)
    string_of_all_texts = authors_text_dictionary[author]
    wordcloud = WordCloud().generate(string_of_all_texts)
    ax.imshow(wordcloud)
    ax.axis('off')

plt.savefig('../figs/word_cloud_by_author')
plt.show()
```

### Finding the Most Commonly Used Tri-Grams

This investigation of the data can be extended into finding most common pairings of words, as this may give us more insight into common thoughts expressed in the text rather than isolated words. Looking into N-Grams (ordered strings of N words) may serve beneficial in exploring common themes.

```python
#create a list of tokenized text (collective) with * indicating separate sentences (to remove from n-gram consideration)
tmp = df['tokenized_txt_array'].to_numpy()
tmp_space = []

for i in range(len(tmp)): 
    tmp_space.append(list('*'))

#create the list
tmp_text = np.array([[i, j] for i, j in zip(tmp, tmp_space)]).ravel()
tmp_arr_of_all_texts = np.concatenate(tmp_text, axis=0)

#build a ngram finder on top of the list using nltk
finder = nltk.collocations.TrigramCollocationFinder.from_words(tmp_arr_of_all_texts)
finder.ngram_fd.most_common(10)

#VISUALIZATION:
top_ngrams = []
i = 0
j = 1

#only the most common 5 tri-grams that don't contain a spacer '*'
while(i < 5): 
    n_gram_list = list(finder.ngram_fd.most_common(j)[-1][0])
    n_gram_count = finder.ngram_fd.most_common(j)[-1][1]
    
    if ('*' in n_gram_list): 
        j += 1
        continue
        
    elif('*' not in n_gram_list): 
        top_ngrams.append([n_gram_list, n_gram_count])
        i += 1 
        j += 1

n_grams_df = pd.DataFrame(top_ngrams, columns =['tri_gram', 'frequency']) 
fig = n_grams_df.plot.bar(x='tri_gram', y="frequency")
fig.get_figure().savefig('../figs/common_tri-grams.png')
```

Not much insight can be derived from this chart. Here, we see that a lot of the most common tri-grams in the text are normal sentence phrases. Based on the results, the top 5 tri-grams are: 

1. "It is not" 
2. "That it is" 
3. "In order to" 
4. "It is the" 
5. "At the same" 

Notably the 3rd most common tri-gram is interesting because a call to action is likely to follow from the phrase "in order to", which might indicate that anticipation or urgency might be a common emotion felt in a lot of these texts. It also indicates that a lot of the texts were written with a specific purpose, to urge something or convey a point.

## 2. Topic Modelling: <a class="anchor" id="second-bullet"></a>

Looking deeper into the topics of interest, I will try to apply the LDA Topic Model learned in class to see if there are any reoccuring themes in the collective texts. 

```python
stop_words = stopwords.words('english')
stop_words.extend(['from', 'subject', 're', 'edu', 'use'])

def sent_to_words(sentences):
    for sentence in sentences:
        # deacc=True removes punctuations
        yield(gensim.utils.simple_preprocess(str(sentence), deacc=True))
        
def remove_stopwords(texts):
    return [[word for word in simple_preprocess(str(doc)) 
             if word not in stop_words] for doc in texts]

df['text_processed'] = df['tokenized_txt'].map(lambda x: x.lstrip('[').rstrip(']').replace("'", "").replace(",", ""))

data = df.text_processed.values.tolist()
data_words = list(sent_to_words(data))
data_words = remove_stopwords(data_words)

# Create Dictionary
id2word = corpora.Dictionary(data_words)
# Create Corpus
texts = data_words
# Term Document Frequency
corpus = [id2word.doc2bow(text) for text in texts]
# View

# number of topics
num_topics = 5

# Build LDA model
lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                       id2word=id2word,
                                       num_topics=num_topics)

# Print the Keyword in the  topics
pprint(lda_model.print_topics())
doc_lda = lda_model[corpus]
```

When 5 categories are assigned to the model, the LDA results in the above, where the coefficient of each word corresponds to its importance/magnitude to the topic. Though not all of them are clear cut, if I had to assign each topic a label it would be something like: 

0. Action/Urgency
1. The Mind
2. Women Workforce, Labour, Economy
3. Knowledge
4. Character

These are all very rough guesses and displays one of the downfalls of LDA. Perhaps experimenting with more/less topics may help narrow down on more solid topics/themes but it is too costly to perform in this exploration.

## 3. Comparing Authors Using Topics and Sentiment <a class="anchor" id="third-bullet"></a>

Based on the exploratory analysis and the results of topic modelling (sections 1 and 2 respectively), it seems that we can class the thoughts expressed in the text in many ways. One of the ways is World vs. Human, where the Human side focuses more on the issues regarding man, and the World side focuses more on the environment, the universe, and so on. Another possible way is to think of people as individuals or as a collective unit, for instance, talking about a person as compared to a society or state. Using these topics, we can plot their topic focus, as well as their sentiment (positive or negative) regarding these topics on a 2D plane to categorize them into 4 major quadrants, and hopefully identify similar thinking authors. 

```python
def human_world(row): 
    
    human = 0
    world = 0
    
    human_words = ['Man', 'Men', 'People', 'Person', 'Human', 'Woman', 'Women', 
                   'Children', 'Ourselves', 'man', 'men', 'people', 'person', 
                   'human', 'woman', 'women', 'children', 'ourselves']
    
    world_words = ['World', 'Universal', 'Animals', 'Country', 'Earth', 'Land', 
                   'Universe', 'Nature', 'world', 'universal', 'animals', 
                   'country', 'earth', 'land', 'universe', 'nature']
    
    sentence = row['tokenized_txt_array']
    
    for term in human_words: 
        human += sentence.count(term)
    
    for term in world_words: 
        world += sentence.count(term)
    
    total = human + world
    
    if total == 0: 
        return {'human' : 0, 'world' : 0}
    
    return {'human' : (human/total), 'world' : (world/total)}

def individual_collective(row): 
    
    individual = 0
    collective = 0
    
    individual_words = ['Man', 'man', 'Woman', 'woman', 'Person', 'person', 'Child', 'child']
    
    collective_words = ['World', 'People', 'Men', 'Human', 'Women', 'People', 'Earth', 'Country', 
                       'Children', 'Society', 'Species', 'Ourselves', 'Government', 'Everyone', 
                       'world', 'people', 'men', 'human', 'women', 'people', 'earth', 'country', 
                       'children', 'society', 'species', 'ourselves', 'government', 'everyone']
    
    sentence = row['tokenized_txt_array']
    
    for term in individual_words: 
        individual += sentence.count(term)
    
    for term in collective_words: 
        collective += sentence.count(term)
    
    total = individual + collective
    
    if total == 0: 
        return {'individual' : 0, 'collective' : 0}
    
    return {'individual' : (individual/total), 'collective' : (collective/total)}
```

```python
emotions_df = pd.read_csv('../output/emotions_df.csv')

def bucket_dates(row): 
    if(row['original_publication_date'] < 1600):
        return '< 1600'
    
    elif(row['original_publication_date'] < 1700):
        return '1600-1700'
    
    elif(row['original_publication_date'] < 1800):
        return '1700-1800'
    
    elif(row['original_publication_date'] < 1900):
        return '1800-1900'
    
    elif(row['original_publication_date'] < 2000):
        return '1900-2000'
    
emotions_df['date bucket'] = emotions_df.apply(lambda row: bucket_dates(row), axis=1)

tmp = emotions_df.groupby('date bucket').mean()[['negative', 'sadness',
       'surprise', 'positive', 'trust', 'anticipation', 'anger', 'fear',
       'disgust', 'joy']]

date = list(tmp.index.values)
date.insert(0, date.pop())

negative = list(tmp['negative'].values)
negative.insert(0, negative.pop())

sadness = list(tmp['sadness'].values)
sadness.insert(0, sadness.pop())

surprise = list(tmp['surprise'].values)
surprise.insert(0, surprise.pop())

positive = list(tmp['positive'].values)
positive.insert(0, positive.pop())

trust = list(tmp['trust'].values)
trust.insert(0, trust.pop())

anticipation = list(tmp['anticipation'].values)
anticipation.insert(0, anticipation.pop())

anger = list(tmp['anger'].values)
anger.insert(0, anger.pop())

fear = list(tmp['fear'].values)
fear.insert(0, fear.pop())

disgust = list(tmp['disgust'].values)
disgust.insert(0, disgust.pop())

joy = list(tmp['joy'].values)
joy.insert(0, joy.pop())

figure(figsize=(20, 20), dpi=80)

emotion_vars = [negative, sadness,
       surprise, positive, trust, anticipation, anger, fear,
       disgust, joy]

emotions = ['negative', 'sadness',
       'surprise', 'positive', 'trust', 'anticipation', 'anger', 'fear',
       'disgust', 'joy']

for i in range(len(emotions)):

    plt.subplot(4, 4, i+1)
    plt.plot(date, emotion_vars[i])
    plt.title(emotions[i])

plt.savefig('../figs/sentiments_over_time.png')
plt.show()
```

Based on the results above there seems to be no clear general trend for any emotion, though most positive emotions seem to be consistently going down more recently. One of the limitations here is that there are also very few points coming from relatively few sources giving the wide time-period intervals. Negative emotions such as negative, sadness, anger, and fear seem to be more unpredictable, and may be related to specific moments in history this would suggest.
